```python
from textblob import TextBlob
import pandas as pd
```

```python
data = pd.read_json("../solrData/clean_reviews.json", orient='records')
data['review_publication_date'] = pd.to_datetime(data['review_publication_date'],unit='ms')
data
```

```python
polarity = []
for review in data['review_content']:
    polarity.append(TextBlob(review).sentiment.polarity)
```

```python
data['polarity'] = pd.DataFrame(polarity)

```

```python
import seaborn as sns

sns.set_style('whitegrid')
sns.displot(data['polarity'], kde = False, color ='red', bins = 40)

data['review_publication_date'] = data['review_publication_date'].astype(str)

data
```

```python
data.to_json('clean_reviews_polarity.json', orient='records')
```

```python
import spacy
import en_core_web_sm
```

```python
polarity = []
nlp = en_core_web_sm.load()
nlp.add_pipe('spacytextblob')

for review in data['review_content']:
    doc = nlp(review)
    polarity.append(doc._.blob.polarity)


```

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.displot(polarity, kde = False, color ='red', bins = 40)
```
