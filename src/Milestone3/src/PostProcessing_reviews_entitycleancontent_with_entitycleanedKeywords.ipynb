```python
import pandas as pd
```

```python
data = pd.read_json("reviews_with_keywords_EmptyContent.json", orient='records')
data = data.drop(columns = ['keywords', 'key_words_with_entities', 'entities', 'review_content'])
print(data.columns)
print(len(list(data.columns)))
```

```python
data
```

```python
data = data.rename(columns= {'empty_content': 'review_content'})
```

```python
data
```

```python
only_words = []
key_words = data['key_words_empty_content']

for i in range(len(key_words)):
    a = key_words[i]
    if not a:
        only_words.append([' '])
        continue
    

    words = list(list(zip(*a))[0])
    only_words.append(words)
```

```python
data['keywords'] = only_words
data = data.drop(columns = 'key_words_empty_content')
```

```python
data.to_json('ProcessedData/reviews_EmptyContent_with_EmptyKeywords.json', orient='records')
```
